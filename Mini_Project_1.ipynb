**Data cleaning**

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
print(df)
```

```python
total_sales = df.iloc[:,1:].sum().sum()

print("Total Sales=",total_sales)
```

**product-wise total sales**

```python
df["Total_Sales"]=df.iloc[:,
1:].sum(axis=1)

month_sales = df.groupby("Month").sum()

print(month_sales)
```

```python
product_sales = df[['Cream','Detergent','Moisturizer','Sanitizer','Shampoo','Soap']].sum()
product_sales.plot(kind="bar")
plt.title("product wise sales")
plt.xlabel("products")
plt.ylabel("sales")
plt.show()
```
