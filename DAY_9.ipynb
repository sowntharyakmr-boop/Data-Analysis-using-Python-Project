```python
import matplotlib.pyplot as plt
expenses = [5000,3000,4000]
categories = ["Food","Travel","Shopping"]
plt.pie(expenses,
        labels=categories)
plt.show()
```

```python
import matplotlib.pyplot as plt
expenses = [5000,3000,4000]
categories = ["Food","Travel","Shopping"]
plt.pie(expenses,
        labels=categories,autopct="%1.1f%%",colors=["Red","Green","Yellow"])
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ["jan","feb","mar","apr","may","jun"]
sales = [10000,12000,15000,10000,13000,20000]
plt.figure(figsize=(12,5));

#Line chart
plt.subplot(1,2,1)
plt.plot(months,sales,marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

#Bar chart
plt.subplot(1,2,2)
plt.bar(months,sales,color="Orange")
plt.title("Sales Comparision")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()

#pie chart
plt.figure(figsize=(6,6))
plt.pie(
    sales,labels=months,
    autopct='1%.1f%%'
)
plt.title("Sales Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
marks = [45,50,55,60,65,78,76,80,85,90.80,78]
plt.hist(marks,bins=5)
plt.title("Student marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
import matplotlib.pyplot as plt
hours = [1,2,3,4,5,6,7,8]
marks = [35,45,58,78,68,76,89,90]
plt.scatter(hours,marks)
plt.title("Study hours vs marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
