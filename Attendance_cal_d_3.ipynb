```python
def calculate_average(a,b,c):
  return(a+b+c)/3

try:
  a=int(input("Enter 1st number:"))
  b=int(input("Enter 2nd number:"))
  c=int(input("Enter 3rd number:"))
  print("calculate_average=",calculate_average(a,b,c))
except ValueError:
  print("Invalid Input")
  print("Invalid Input")
```
