```python
import pandas as pd
data = {
    "Region":["South","North","East","West","South","North"],
    "Products":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,40000,350000,250000,200000,300000]
}
df = pd.DataFrame(data)
print(df)
```

```python
Region_sales = df.groupby("Region")["Sales"].sum()

print(Region_sales)
```

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum","mean","max","min"]
)

print(result)
```

```python
south_sales = df[df["Region"]=="South"]

print(south_sales)
```

```python
high_sales = df[df["Sales"]> 250000]

print(high_sales)
```

```python
import pandas as pd

data = {
    'Product':['Laptop','mobile','None','Tablet','Laptop'],
    'Sales':[50000,30000,20000,None,50000],
    'Quantity':[5,3,None,2,5]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

df = df.drop_duplicates()

df['Products'] = df['Product'].fillna('Unknown')

df['Sales'] = df['Sales'].fillna(df['Sales'].mean())

df['Quantity'] = df['Quantity'].fillna(0)

print("\nCleaned Dataset:")
print(df)
```
