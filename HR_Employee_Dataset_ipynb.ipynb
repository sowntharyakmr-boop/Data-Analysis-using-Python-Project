**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the Dataset**

```python
import pandas as pd
df = pd.read_csv("employee_attrition.csv")
print(df.head())
```

**Data Exploration**

```python
print(df.info())
```

**Clean Employee Dataset**

```python
#Missing values check
print(df.isnull().sum())

#Remove missing values
df = df.dropna()

#Remove duplicates
df = df.drop_duplicates()

print(df.shape)
df.head()
```

**Missing Values Check**


```python
print(df.isnull().sum())
```

**Employee Attrition Counts**

```python
print(df['Attrition'].value_counts())
```

**Attrition Pie Chart**

```python
import matplotlib.pyplot as plt
df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Employee Attrition Rate")
plt.show()
```

**Department-wise Attrition**

```python
pd.crosstab(df['Department'],
df['Attrition'])
```

**Department-wise Bar chart**

```python
pd.crosstab(df['Department'],
df['Attrition']).plot(kind="bar")
plt.title("Department Wise Attrition")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

**Salary Analysis**

```python
print(df.groupby('Attrition')
['Salary'].mean())
```

**Experience Analysis**

```python
print(df.groupby('Attrition')
['Experience'].mean())
```

**Salary chart**

```python
plt.bar(df['Department'],df['Salary'])
plt.title("Department vs Salary")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

**Experience Chart**

```python
plt.bar(df['Department'],df['Experience'])
plt.title("Department vs Experience")
plt.xlabel("Department")
plt.ylabel("Experience")
plt.show()
```
